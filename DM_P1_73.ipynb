```python
#Remove Outliers Using IQR:
# Calculate the Interquartile Range (IQR)
Q1 = df['age'].quantile(0.25)
Q3 = df['age'].quantile(0.75)
IQR = Q3 - Q1
```

```python
# prompt: explain the following code in detail df = df[~((df['age'] < (Q1 - 1.5 * IQR)) | (df['age'] > (Q3 + 1.5 * IQR)))]

# This line of code removes outliers from a Pandas DataFrame 'df' based on the 'age' column using the Interquartile Range (IQR) method.
# Let's break it down step by step:

# 1. `df['age']`: This selects the 'age' column from the DataFrame 'df'.

# 2. `(df['age'] < (Q1 - 1.5 * IQR))`: This part of the code checks if the 'age' values are below the lower limit of the IQR range.
#   - `Q1`: The first quartile (25th percentile) of the 'age' column.
#   - `IQR`: The interquartile range (Q3 - Q1).
#   - `1.5 * IQR`: This is a common multiplier used in the IQR method to identify outliers.
#   - `Q1 - 1.5 * IQR`: The lower limit of the IQR range.
#   - `df['age'] < (Q1 - 1.5 * IQR)`: This results in a boolean Series where True indicates values below the lower limit.

# 3. `(df['age'] > (Q3 + 1.5 * IQR))`: This part of the code checks if the 'age' values are above the upper limit of the IQR range.
#   - `Q3`: The third quartile (75th percentile) of the 'age' column.
#   - `Q3 + 1.5 * IQR`: The upper limit of the IQR range.
#   - `df['age'] > (Q3 + 1.5 * IQR)`: This results in a boolean Series where True indicates values above the upper limit.

# 4. `((df['age'] < (Q1 - 1.5 * IQR)) | (df['age'] > (Q3 + 1.5 * IQR)))`: This combines the two boolean Series using the '|' (OR) operator.
#   - This results in a boolean Series where True indicates values that are either below the lower limit or above the upper limit of the IQR range.
#   - Essentially, it identifies outliers.

# 5. `~(...)`: This negates the boolean Series.
#   - It effectively selects only the rows where the 'age' values are within the IQR range (i.e., not outliers).

# 6. `df[...]`: This selects only the rows from the DataFrame 'df' where the corresponding boolean value is True.

# In summary, the line of code `df = df[~((df['age'] < (Q1 - 1.5 * IQR)) | (df['age'] > (Q3 + 1.5 * IQR)))]` removes the rows from the DataFrame 'df' where the 'age' values are considered outliers based on the IQR method.
```

```python
df = df[~((df['age'] < (Q1 - 1.5 * IQR)) | (df['age'] > (Q3 + 1.5 * IQR)))]
```

```python
# Convert all categorical values to lowercase for standardization
df['rbc'] = df['rbc'].str.lower()
# Replace inconsistent values
df['rbc'].replace({'n': 'normal', 'ab': 'abnormal'}, inplace=True) #df
```

```python
# Validation check: Age should be between 0 and 100
invalid_age = df[(df['age'] < 0) | (df['age'] > 100)]
if not invalid_age.empty:
  print("Invalid age entries found:\n", invalid_age)
# Validation check for blood pressure values
invalid_bp = df[(df['bp'] < 0) | (df['bp'] > 300)]
if not invalid_bp.empty:
  print("Invalid blood pressure entries found:\n", invalid_bp)
```
